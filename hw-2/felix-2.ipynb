```python
# a) Load the dataset, center & make scatter plot

# load data
Data = np.genfromtxt("pca-data-2d.dat.txt", skip_header=False,dtype='float')

# center data
mean = np.mean(Data)
centered = Data - mean
plt.scatter(centered[:,0],centered[:,1])
plt.show()
```

```python
# b) Determine the Principal Component Directions (PCs) and make another scatter plot of the
# same data points in the coordinate system spanned by the 2 PCs.

# 1. Cov Matrix
C = np.cov(centered.T)
# 2. get eigenvectors and eigenvalues of cov matrix, w values, v vectors
w,v = np.linalg.eigh(C)
v1 = np.asmatrix(v[:,0])
v2 = np.asmatrix(v[:,1])

# 3.plot data points in the coordinate systems spanned by the 2 PCs.
```

```python
# c)Plot the reconstruction of the data in the original coordinate system when using 
# (i) only the first or 
rec_1 = v1.dot(centered.T)
plt.plot(rec_1.T)
plt.show()

# (ii) only the second PC for reconstruction.
rec_2 = v2.dot(centered.T)
plt.plot(rec_2.T)
plt.show()
```

# 2.2 PCA: 3-dimensional Toy Data

```python
# a) Load the dataset, center & make scatter plot

# load data
Data = np.genfromtxt("pca-data-3d.txt",delimiter=',', skip_header=True,dtype='float')
# center data
mean = np.mean(Data)
centered3d = Data - mean

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(centered3d[:,0], centered3d[:,1], zs=centered3d[:,2])
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```

```python
#(b) Determine the PCs and make the analogous scatterplot matrix for the 2d-coordinate systems
# spanned by the different pairs of PCs.

# 1. Cov Matrix
C3d = np.cov(centered3d.T)
# 2. get eigenvectors and eigenvalues of cov matrix, w values, v vectors
w3d,v3d = np.linalg.eigh(C3d)
v13d = np.asmatrix(v3d[:,0])
v23d = np.asmatrix(v3d[:,1])
v33d = np.asmatrix(v3d[:,2])
```

```python
# Examine the 3d-reconstruction of the data in the original coordinate systems when using only 
# (i) the first, 
rec_1_3d = v13d.dot(centered3d.T)
plt.plot(rec_1_3d.T)
plt.show()

#(ii) the first two or 
rec_2_3d = v23d.T.dot(rec_1_3d)
plt.scatter(rec_2_3d[0,:],rec_2_3d[1,:])
plt.show()

#(iii) all three PCs for reconstruction. 
rec_3_3d = v33d.dot(rec_2_3d)
#print (rec_3_3d.shape)

fig1 = plt.figure()
ax1 = Axes3D(fig1)
ax1.scatter(rec_3_3d[0,:], rec_3_3d[1,:], zs=rec_3_3d[2,:])
ax1.set_xlabel('X Label')
ax1.set_ylabel('Y Label')
ax1.set_zlabel('Z Label')
plt.show()

# Discuss how useful these directions (i.e., the PCs) are
```
