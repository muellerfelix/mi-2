```python
% matplotlib inline

import numpy as np
import scipy as sp
import math
import scipy.stats as scp
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from PIL import Image
import numpy as np

%matplotlib inline
```

```python
# load data
Data = np.genfromtxt("expDat.txt", delimiter=',', skip_header=True,dtype='float')
# skip first row ("1","2",..)
X = Data[:,1:]
```

```python
m = np.mean(np.mean(X))
d_center = np.subtract(X, m)
cov = (1./100.)*np.dot(d_center,d_center.T)


plt.imshow(cov, cmap='hot', interpolation='nearest')
plt.title("selfmade")
plt.show()


plt.imshow(np.cov(X), cmap='hot', interpolation='nearest')
plt.title("numpy")
plt.show()
```
